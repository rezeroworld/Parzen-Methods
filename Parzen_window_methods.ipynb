## Random Projections
One method to reduce the dimensionality of the data (eventhough we have low dimensionality data) is to use random projections implemented in solution.py, the idea is to use a random linear transformation to reduce the data to dimension 2, the transformation is a random variable sampled from a normal distribution with mean 0 and variance 1. Let's see how the algorithms perform after applying the linear transformation

```python
solution.get_test_errors_projections(data, with_plot=True, number_of_projections=500, with_e_bars=True)
```

This plot shows the average error rate over 500 random projections, in addition to not having brought any benefit to the performance the linear transformation decreased the performance of both algorithms. This leads to the conclusion that the data are not linearly separable, but to be sure of that one needs to save the best performing transformation and analyze it.
